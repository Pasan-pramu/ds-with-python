# Loops


```python
# Example 1: Loop through a list
numbers = [10, 20, 30]
for n in numbers:
    print(n)

# Example 2: Loop a fixed number of times
for i in range(5):
    print("Count:", i)
```

# List Comprehensions

```python
# Example 1: Square every number
numbers = [1, 2, 3, 4, 5]
squares = [n * n for n in numbers]
print(squares)        # [1, 4, 9, 16, 25]

# Example 2: Keep only even numbers (with a condition)
numbers = [1, 2, 3, 4, 5, 6]
evens = [n for n in numbers if n % 2 == 0]
print(evens)          # [2, 4, 6]
```

# Functions

```python
# Example 1: Simple function with a return
def add(a, b):
    return a + b

print(add(5, 3))      # 8

# Example 2: Function with a default value
def greet(name="Student"):
    return f"Hello, {name}!"

print(greet())            # Hello, Student!
print(greet("Kasun"))     # Hello, Kasun!
```
