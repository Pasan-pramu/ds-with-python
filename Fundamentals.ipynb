```python
# Create a dictionary
student = {"name": "Kasun", "age": 22, "city": "Kandy"}
print(student)
```

```python
# Get a value by key
print(student["name"])
```

```python
# Safer way to get a value
print(student.get("age"))
```

```python
# Add a new key
student["course"] = "Data Science"
print(student)
```

```python
# Change a value
student["age"] = 23
print(student)
```

```python
# Delete a key
del student["city"]
print(student)
```

```python
# Get all keys
print(student.keys())
```

```python
# Get all values
print(student.values())
```

```python
# Get key-value pairs
print(student.items())
```

```python
# Check if a key exists
print("name" in student)
```

```python
# Check if a key exists
print("name" in student)
```

```python
# Length (how many items)
print(len(student))
```
